```python
def backtest(weather, model, predictors, start=1825, step=30):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        # Set up training data
        train = weather.iloc[:i, :]
        # Set up test data
        test = weather.iloc[i:(i + step), :]

        # Fit the model on the training data
        model.fit(train[predictors], train["target"])

        # Predict on the test set
        preds = model.predict(test[predictors])  # This returns a numpy array
        preds = pd.Series(preds, index=test.index)  # Convert array to pandas Series with test indices

        # Combine actual values and predictions
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        # Append the combined DataFrame to the list
        all_predictions.append(combined)

    # Concatenate all predictions into a single DataFrame
    return pd.concat(all_predictions)
```

```python
predictions=backtest(weather, rr,predictors)
```

```python
predictions
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(predictions["actual"],predictions["prediction"])
```

```python
def pct_diff( old, new): #to calculate percentage difference
    return(new-old)/old

def compute_rolling(weather,horizon,col): #horizon=no. of days we want to compute rolling average for and col= on which col
    label=f"rolling_{horizon}_{col}" #just the name of the column which we will make acc.
    weather[label]=weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"]=pct_diff(weather[label],weather[col])
    return weather

rolling_horizons=[3,14]

for horizon in rolling_horizons:
    for col in["BASEL_temp_max","BASEL_temp_min","BASEL_precipitation"]:
        weather=compute_rolling(weather,horizon,col)
```

```python
def expand_mean(df):
    return df.expanding(1).mean()

for col in["BASEL_temp_max","BASEL_temp_min","BASEL_precipitation"]:
    weather[f"month_avg_{col}"]=weather[col].groupby(weather.index.month,group_keys=False).apply(expand_mean)
    weather[f"day_avg_{col}"]=weather[col].groupby(weather.index.day_of_year,group_keys=False).apply(expand_mean)
```

```python
import warnings

try:
    with warnings.catch_warnings():
        warnings.simplefilter("error")
        predictions = backtest(weather, rr, predictors)
except Warning as w:
    print("A warning occurred during backtesting:")
    print(w)
except Exception as e:
    print("An error occurred while running the backtest:")
    print(e)
```
